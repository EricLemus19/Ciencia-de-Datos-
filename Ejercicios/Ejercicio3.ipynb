# Ejercicio **3**

```python
# Creando la matriz B 
B = np.array(np.arange(0.4,0.8,0.01)).reshape(8,5)
B
```

```python
# Concatenar las matrices
N = np.concatenate((M,B), axis = 1)
N.shape
```

```python
np.mean(N, axis=0)
```

```python
# Parte tres del ejercicio
Z = np.append(N,[np.mean(N, axis=0)], axis= 0)
Z.shape
```

```python
Z
```

```python
# Parte 4. Agregar una columna al final con el valor maximo de cada fila
F = np.max(Z, axis=1)
data = np.concatenate((Z,F[:,None]), axis = 1)
print(data.shape)

# Parte 5. Guardar el archivo 
np.savetxt('datos.txt', data, delimiter=',', fmt='%.3f')
```

```python
import matplotlib.pyplot as plt
from PIL import Image
 
imagen = Image.open('/content/Full_Moon_Luc_Viatour.jpg')
imagen.mode
pixeles = imagen.load()
largo, ancho = imagen.size

for x in range(largo):
    for y in range(ancho):
          (r,g,b) = pixeles[x, y]
          if (r, g, b) == (0,0,0):
            pixeles[x, y] = (255, 255, 200)

fig,ax = plt.subplots(figsize=(6,6))
ax.imshow(imagen)
plt.show()
```

# **Ejercicio 4**

```python
import numpy as np
imagen = Image.open("/content/Capivara.jpg")
im = np.array(imagen)

for y in range(im.shape[0]):
    for x in range(im.shape[1]):
      pixel = im[y][x]
      promedio = (pixel[0] + pixel[1] + pixel[2])//3
      im[y][x][0] = promedio
      im[y][x][1] = promedio
      im[y][x][2] = promedio

fig_2,ax_2 = plt.subplots(figsize=(6,6))
ax_2.imshow(im)
plt.show()
```

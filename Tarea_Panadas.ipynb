```python
# Función que grafica las columnas de interes para paises que pertenecen al G20 
def G20(Pais,df): 
  pais = str(Pais)
  if pais in df['Country'].values:
    print(f'Sí, {pais} está en la lista de países del G20.')
    df_pais = df.loc[df['Country'] == Pais]
    fig = plt.figure()
    axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
    axes.grid()
    axes.set_title(pais)
    axes.plot(df_pais['Total'])
    axes.plot(df_pais['Oil'])
    axes.plot(df_pais['Coal'])
    axes.plot(df_pais['Gas'])
    axes.plot(df_pais['Cement'])
    axes.plot(df_pais['Flaring'])
    axes.plot(df_pais['Per Capita'])
    axes.legend(['Total','Oil','Coal','Gas', 'Cement', 'Flaring', 'Per Capita'])
  else:
    print(f'Lo siento, {pais} no está en la lista de países del G20.')

G20('Argentina',df)
```

```python
# Ver si los paises se encuentran en el DataFrame
g20 = ['Argentina', 'Australia', 'Brazil', 'Canada', 'China', 'France', 'Germany', 'India', 'Indonesia', 'Italy', 'Japan', 'Mexico', 'Russia', 'Saudi Arabia', 'South Africa', 'South Korea', 'Turkey', 'United Kingdom', 'USA', 'European Union']
lista = [True if item in df['Country'].values else False for item in g20]
np.array(lista).T
```

```python
for value in g20:
  G20(value,df)
```

```python
# Función que nos dice que pais contamina más para la categoría que se elija
def Mayor(categoria,df):
  df4 = pd.DataFrame(df.groupby(['Country'])[categoria].max())
  top_20 = df4.nlargest(5, categoria)
  return print(top_20[1:-1])

lis = ['Total','Oil','Coal','Gas', 'Cement', 'Flaring']
[Mayor(value,df) for value in lis]
```

```python
# DataFrame de G20
df_g20 =pd.DataFrame()
for value in g20:
  df_aux = df.loc[df['Country'] == value]
  df_g20 = pd.concat([df_g20, df_aux], ignore_index=True)

df_g20
```

```python
# Porcentaje con el que contribuye México al Total y Per Capita
emisiones_totales = df_g20['Total'].sum()
mexico_emisiones_totales = df_g20.loc[df_g20['Country'] == 'Mexico','Total'].sum()
porcentaje_total = (mexico_emisiones_totales / emisiones_totales) * 100

per_totales = df_g20['Total'].sum()
mexico_per_totales = df_g20.loc[df_g20['Country'] == 'Mexico','Per Capita'].sum()
porcentaje_total_per = (mexico_per_totales / per_totales) * 100
print(f'El porcentaje con el que contribuye México al total es: {porcentaje_total} y al Per Capita es: {porcentaje_total_per}')
```

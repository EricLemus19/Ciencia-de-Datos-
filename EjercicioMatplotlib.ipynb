```python
#Ejercicio en clase

import numpy as np
import matplotlib.pyplot as plt

# Funciones
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

def f2(t,t_0,delta):
  return np.exp(-((t-t_0)**2)/(2*delta) + t_0**2/(2*delta))


t  = np.random.randint(0,10,100) 
t1 = np.arange(0.0, 10.0, 0.1)
t2 = np.linspace(-10,10,1000)

plt.figure(figsize=(10,10))
ax1 = plt.subplot(3,3,(1,3))
ax1.plot(t2, f(t2), color = 'orange')
#ax1.margins(2, 2)
ax1.grid(True)


ax2 = plt.subplot(3,3,(4,5))
y =  10*t1 + np.random.randn(100)
ax2.scatter(t1, y, c = t, cmap = 'Blues')
ax2.plot(t1,10*t1 + 1, label = '$10t + 1$')
ax2.legend()
ax2.grid()

ax3 = plt.subplot(3,3,7)
ax3.plot(t2, f2(t2,0,2))
ax3.grid()

ax4 = plt.subplot(3,3,8)
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']
ax4.bar(fruits, counts, label=bar_labels, color=bar_colors)

#ax4.set_ylabel('fruit supply')
#ax4.set_title('Fruit supply by kind and color')
#ax4.legend(title='Fruit color')


ax5 = plt.subplot(3,3,(6,9))
ax5.plot(t2,f2(t2,1,0.5), '-.', lw=3)
ax5.grid()

axes = [ax1, ax2, ax3, ax4, ax5]
plt.show()
```
